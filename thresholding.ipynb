```python
import cv2
from skimage import io
from matplotlib.pyplot import figure
image = cv2.imread('./images/tt.jpg')

# Apply thresholds
img1 = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY)[1]
img2 = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY_INV)[1]

figure(0)
io.imshow(img1)
figure(1)
io.imshow(img2)
```


## Calculating gradient

```python
import cv2
from skimage import io

src = cv2.imread('./images/tt.jpg')

cv2.GaussianBlur(src, (3, 3), 0)

gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

# Apply Sobel method to the grayscale image 
grad_x = cv2.Sobel(gray, cv2.CV_16S, 1, 0, ksize=3, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
grad_y = cv2.Sobel(gray, cv2.CV_16S, 0, 1, ksize=3, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)

abs_grad_x = cv2.convertScaleAbs(grad_x)
abs_grad_y = cv2.convertScaleAbs(grad_y)
grad = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)

io.imshow(grad)
```

```python
from matplotlib.pyplot import figure

src_eqlzd = cv2.equalizeHist(gray)
figure(0)
io.imshow(gray)
figure(1)
io.imshow(src_eqlzd)
```
