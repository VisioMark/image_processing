```python
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

img = cv2.imread('./images/001.jpg')
width= 1162
height = 2000

resized = cv2.resize(img,( width, height), interpolation=cv2.INTER_AREA)

a = cv2.threshold(cv2.cvtColor(cv2.bitwise_not(resized), cv2.COLOR_BGR2GRAY), 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
              
figure(figsize=(10,10), dpi=90)
plt.imshow(a)
```

```python
gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
figure(figsize=(20,20), dpi=90)
plt.imshow(gray, cmap='gray')
```

```python
blurred = cv2.GaussianBlur(gray, (11, 9), cv2.BORDER_DEFAULT)

figure(figsize=(20,20), dpi=90)
plt.imshow(blurred, cmap='gray')
```

```python
thresh_blurred = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]


figure(figsize=(20,20), dpi=90)
plt.imshow(thresh_blurred, cmap='gray')
```
