```python
import cv2
import pytesseract
from skimage import io
from matplotlib.pyplot import figure

img = cv2.imread('./images/download.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

figure(0)
io.imshow(gray)
figure(1)
io.imshow(thresh)
```

```python
import pytesseract

this = cv2.imread('./images/1.jpg')

# Get text from image
text = pytesseract.image_to_string(this, lang='eng', config='--psm 6')
print(text)
```
